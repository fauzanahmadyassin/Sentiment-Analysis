```python
import sys,tweepy,csv,re
import pandas as pd
import numpy as np

pd.set_option('display.notebook_repr_html', True)

class SamplingDataset:

    def MainProcess(self):
        
        df = pd.DataFrame()
        df_corpus = pd.DataFrame()
        
        #############################  DATA BEFORE RESAMPLING
        df = pd.read_csv('CarReview.csv')
        #############################
        df = df.sample(frac=1).reset_index(drop=True)
        display (df)
        
        df= self.Sampling(df)
        
        display (df.head(40))
        
        ####################################################################  SAVE FILE AFTER RESAMPLING
        #df.to_csv('car_resampling.csv', sep=',', index=False, encoding='utf-8')
        ####################################################################
        
    def Sampling(self, df):
        
        df_pos= pd.DataFrame()
        df_neg = pd.DataFrame()
        df_train = pd.DataFrame()
        
        df_neg = df.loc[df['Polarity_Class'] == 'Negative']
        df_neg= df_neg.sample(frac=1).reset_index(drop=True)
        
        df_pos = df.loc[df['Polarity_Class'] == 'Positive']
        df_pos = df_pos.sample(frac=1).reset_index(drop=True)
        
        #display (df_pos)
        #display (df_neg)
        
        print ('\nSize of Positive tweets : '+str(len(df_pos)))
        print ('Size of Negative tweets : '+str(len(df_neg)))
        
        if len(df_pos)>len(df_neg):
            sampling_size = len(df_neg)
            print ('Resample is required from Positive datasets')
        else:
            sampling_size = len(df_pos)
            print ('Resample is required from Negative datasets')
            
        print ('Sampling size is ' +str(sampling_size*2)+ ' where each Positive and Negative tweets are set equally with ' +str(sampling_size)+ ' number of dataset')    
        
        df_pos = df_pos.sample(n=sampling_size)
        df_pos = df_pos.sample(frac=1).reset_index(drop=True)

        train = [df_pos, df_neg]
        df_train = pd.concat(train)
        df_train = df_train.sample(frac=1).reset_index(drop=True)
        
        return df_train
             

if __name__== "__main__":
    sa = SamplingDataset()
    sa.MainProcess() 
    
```
