```python
import sys,tweepy,csv,re
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import matplotlib as mpl
import pylab as plb
import pandas as pd
import numpy as np

pd.set_option('display.notebook_repr_html', True)

class Visualize:

    def MainProcess(self):
        
        df_cars = pd.DataFrame()
        df_honda = pd.DataFrame()
        df_toyota = pd.DataFrame()
        df_hyundai = pd.DataFrame()
        df_peugeot = pd.DataFrame()
        df_audi = pd.DataFrame()
        
        honda_pos = pd.DataFrame()
        honda_neg = pd.DataFrame()
        toyota_pos = pd.DataFrame()
        toyota_neg = pd.DataFrame()
        hyundai_pos = pd.DataFrame()
        hyundai_neg = pd.DataFrame()
        peugeot_pos = pd.DataFrame()
        peugeot_neg = pd.DataFrame()
        audi_pos = pd.DataFrame()
        audi_neg = pd.DataFrame()
        cars_pos = pd.DataFrame()
        cars_neg = pd.DataFrame()

        df_cars = pd.read_csv('CarReview.csv')
        df_honda = pd.read_csv('honda.csv')
        df_toyota = pd.read_csv('toyota.csv')
        df_hyundai = pd.read_csv('hyundai.csv')
        df_peugeot = pd.read_csv('peugeot.csv')
        df_audi = pd.read_csv('audi.csv')
        
        honda_pos = df_honda.loc[df_honda['Polarity_Class'] == 'Positive']
        total_pos_honda = len(honda_pos)
        honda_neg = df_honda.loc[df_honda['Polarity_Class'] == 'Negative']
        total_neg_honda = len(honda_neg)
        
        toyota_pos = df_toyota.loc[df_toyota['Polarity_Class'] == 'Positive']
        total_pos_toyota = len(toyota_pos)
        toyota_neg = df_toyota.loc[df_toyota['Polarity_Class'] == 'Negative']
        total_neg_toyota = len(toyota_neg)
        
        hyundai_pos = df_hyundai.loc[df_hyundai['Polarity_Class'] == 'Positive']
        total_pos_hyundai = len(hyundai_pos)
        hyundai_neg = df_hyundai.loc[df_hyundai['Polarity_Class'] == 'Negative']
        total_neg_hyundai = len(hyundai_neg)
        
        peugeot_pos = df_peugeot.loc[df_peugeot['Polarity_Class'] == 'Positive']
        total_pos_peugeot = len(peugeot_pos)
        peugeot_neg = df_peugeot.loc[df_peugeot['Polarity_Class'] == 'Negative']
        total_neg_peugeot = len(peugeot_neg)
        
        audi_pos = df_audi.loc[df_audi['Polarity_Class'] == 'Positive']
        total_pos_audi = len(audi_pos)
        audi_neg = df_audi.loc[df_audi['Polarity_Class'] == 'Negative']
        total_neg_audi = len(audi_neg)
        
        cars_pos = df_cars.loc[df_cars['Polarity_Class'] == 'Positive']
        total_pos_cars = len(cars_pos)
        cars_neg = df_cars.loc[df_cars['Polarity_Class'] == 'Negative']
        total_neg_cars = len(cars_neg)
        
        print ('\nHonda :')
        print ('Total positive tweet : ' +str(total_pos_honda))
        print ('Total negative tweet : ' +str(total_neg_honda))
               
        print ('\nToyota :')
        print ('Total positive tweet : ' +str(total_pos_toyota))
        print ('Total negative tweet : ' +str(total_neg_toyota))
        
        print ('\nHyundai :')
        print ('Total positive tweet : ' +str(total_pos_hyundai))
        print ('Total negative tweet : ' +str(total_neg_hyundai))
        
        print ('\nPeugeot :')
        print ('Total positive tweet : ' +str(total_pos_peugeot))
        print ('Total negative tweet : ' +str(total_neg_peugeot))
               
        print ('\nAudi :')
        print ('Total positive tweet : ' +str(total_pos_audi))
        print ('Total negative tweet : ' +str(total_neg_audi))

        print ('\nCar :')
        print ('Total positive tweet : ' +str(total_pos_cars))
        print ('Total negative tweet : ' +str(total_neg_cars))
        print ('\n\n')
        
        ########################################################################################################
        # histogram
        n_groups = 5
        pos_bar = (total_pos_honda,total_pos_toyota,total_pos_hyundai,total_pos_peugeot,total_pos_audi)
        neg_bar = (total_neg_honda,total_neg_toyota,total_neg_hyundai,total_neg_peugeot,total_neg_audi)

        # create plot
        index = np.arange(n_groups)
        bar_width = 0.4
        opacity = 0.8
        plt.figure(dpi=150, figsize=(13,9))
        plb.rcParams['font.size'] = 16

        rects1 = plt.bar(index, pos_bar, bar_width,alpha=opacity,color='b',align='edge',label='Positive')
        rects2 = plt.bar(index + bar_width, neg_bar, bar_width,alpha=opacity,color='r',align='edge',label='Negative')

        plt.xlabel('\nCar Brands')
        plt.ylabel('Number of Datasets\n')
        plt.title('Total Datasets by Brands\n')
        plt.xticks(index + bar_width, ('Honda City', 'Toyota Vios', 'Hyundai Accent', 'Peugeot 308', 'Audi A3'))
        plt.legend()
        plt.tight_layout()
        
        self.autolabel(rects1)
        self.autolabel(rects2)
        
        plt.show()
        ########################################################################################################
        
        #pie chart
        print ('\n\n')
        labels = '\nPositive', 'Negative\n'
        sizes = [total_pos_cars,total_neg_cars]
        colors = ['b', 'r']
        explode = (0, 0)  # explode 1st slice

        # Plot
        plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', startangle=190)
        plt.title('   Total of Positive and Negative tweets from all brands\n\n')
        plt.axis('equal')
        plt.show()

        
        
    def autolabel(self,rects):
        for rect in rects:
            height = rect.get_height()
            plt.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                    '%d' % int(height),
                    ha='center', va='bottom')
        
        
        
        
if __name__== "__main__":
    sa = Visualize()
    sa.MainProcess() 


```
